```python
import json
import os

db_file = "kassite_names_db.json"

if os.path.exists(db_file):
    with open(db_file, "r", encoding="utf-8") as file:
        kassite_db = json.load(file)
    print(f"Database loaded! {len(kassite_db)} entries found.")
else:
    kassite_db = []
    print("New database initialized.")
```

```python
import pandas as pd
import os


excel_file = None
for f in os.listdir():
    if f.startswith("Kassitische") and f.endswith(".xlsx") and not f.startswith("~$"):
        excel_file = f
        break

if excel_file:
    print(f"✅ Datei gefunden und wird geladen: {excel_file}")
    df = pd.read_excel(excel_file)
    
    df.columns = df.columns.str.strip()

    for _, row in df.iterrows():

        neu_eintrag = { 
            "transcription": str(row["Name"]), 
            "morphemes": str(row["Morpheme"]) if pd.notna(row["Morpheme"]) else "",
            "meaning": str(row["Bedeutung"]) if pd.notna(row["Bedeutung"]) else "",
            "position": str(row["Position"]) if pd.notna(row["Position"]) else "",
            "function": str(row["Funktion"]) if pd.notna(row["Funktion"]) else ""
        }
        

        kassitische_datenbank.append(neu_eintrag)

    print(f"✅ Erweitert! Die Datenbank hat jetzt {len(kassitische_datenbank)} Einträge.")
else:
    print(f"❌ Excel-Datei '{excel_file}' nicht gefunden. Liegt sie im selben Ordner?")

with open(db_file, "w", encoding="utf8") as f:
    json.dump(kassite_db, f, indent=4, ensure_ascii=False)

print(f"✅ Alles erfolgreich in {db_file} gespeichert!")

def sync_to_github() :
    os.system("git add .")
    
    commit_msg = "Update database from Excel import"
    if os.system(f'git commit -m "{commit_msg}"') == 0:
        os.system("git push origin main")
        print("Synchronisation mit GitHub abgeschlossen!")
    else:
        print("Keine Änderungen zum Hochladen gefunden.")

sync_to_github
```
